```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

names=[]
rating=[]
reviews=[]
reviewsValue=[]
interLink=[]
pages=30
head={
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.4; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Referer': 'https://www.ambitionbox.com/',
}

for pg in range(1,pages+1):
    url=f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={pg}"
    response=requests.get(url,headers=head)
    if response.status_code==200:
        print(f'Successfully Retrived : {pg}')
        soup=BeautifulSoup(response.content,'lxml')
        
        for i in soup.find_all('h2'):
            name=i.text.strip()
            if name:
                names.append(name)
        for span in soup.find_all('span',class_='companyCardWrapper__companyRatingValue'):
            value=span.text.strip()
            if value:
                rating.append(value)
            else:
                print(f'some problem to retriving rating values')
        for j in soup.find_all('span',class_='companyCardWrapper__ActionTitle'):
            rev=j.text.strip()
            if rev:
                reviews.append(rev)
            else:
                print(f'problem to retrive Reviews List')
        for k in soup.find_all('span',class_='companyCardWrapper__ActionCount'):
            counts=k.text.strip()
            if counts:
                reviewsValue.append(counts)
            else:
                print(f'problem to retrive Review Action Count List')
        for m in soup.find_all('span',class_='companyCardWrapper__interLinking'):
            links=m.text.strip()
            if links:
                interLink.append(links)
            else:
                print(f'problem to retrive interLink ')
            
            
        
    else:
        print(f'Failed to retrived {pg}')
        
Series1=pd.Series(names, name='CompanyNames')
Series2=pd.Series(rating,name='Rating')
Series3=pd.Series(reviews,name='Reviews')
Series4=pd.Series(reviewsValue,name='ActionCount')
        


        
```
